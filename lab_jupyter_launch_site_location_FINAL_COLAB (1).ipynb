```python
# Task 2: Mark the success/failed launches for each site on the map

import folium
from IPython.display import display

# Create a fresh map centered around the average coordinates
launch_map = folium.Map(location=[spacex_df['Lat'].mean(), spacex_df['Long'].mean()], zoom_start=5)

# Loop through all launch records and mark each with color-coded circle
for index, row in spacex_df.iterrows():
    lat = row['Lat']
    lon = row['Long']
    site = row['Launch Site']
    outcome = row['class']

    # Green = Success, Red = Failure
    color = 'green' if outcome == 1 else 'red'
    outcome_label = 'Success' if outcome == 1 else 'Failure'

    folium.CircleMarker(
        location=[lat, lon],
        radius=4,
        color=color,
        fill=True,
        fill_color=color,
        fill_opacity=0.7,
        popup=f"{site} - {outcome_label}"
    ).add_to(launch_map)

# Display map
display(launch_map)

```

```python
# ✅ Completed: Haversine distance calculation (no geopy needed)
from math import radians, cos, sin, asin, sqrt
def haversine_distance(coord1, coord2):
    lon1, lat1 = coord1[1], coord1[0]
    lon2, lat2 = coord2[1], coord2[0]
    lon1, lat1, lon2, lat2 = map(radians, [lon1, lat1, lon2, lat2])
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    c = 2 * asin(sqrt(a))
    km = 6371 * c
    return km

# Example distances for CCAFS LC-40
launch_site = [28.562302, -80.577356]
coastline = [28.563197, -80.567888]
highway = [28.562951, -80.570845]
railway = [28.572057, -80.585278]
city = [28.6100, -80.8076]
distance_coast = haversine_distance(launch_site, coastline)
distance_highway = haversine_distance(launch_site, highway)
distance_railway = haversine_distance(launch_site, railway)
distance_city = haversine_distance(launch_site, city)
print(f"Coastline: {distance_coast:.2f} km")
print(f"Highway: {distance_highway:.2f} km")
print(f"Railway: {distance_railway:.2f} km")
print(f"City: {distance_city:.2f} km")
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
import folium
from folium import Marker, PolyLine

# Recreate base map
site_map = folium.Map(location=[28.562302, -80.577356], zoom_start=6)

# Example: Kennedy Space Center
launch_site = [28.562302, -80.577356]

# Example proximity points (replace with real proximity data if you have)
nearest_city = [28.385233, -81.563874]      # Orlando
nearest_highway = [28.558611, -80.810556]   # Near Route 405
nearest_railway = [28.572872, -80.585556]   # Railway nearby
nearest_coastline = [28.563197, -80.567888]

# List of proximities with labels
proximities = {
    'City': nearest_city,
    'Highway': nearest_highway,
    'Railway': nearest_railway,
    'Coastline': nearest_coastline
}

# Plot Launch Site
Marker(launch_site, tooltip='Launch Site', icon=folium.Icon(color='blue')).add_to(site_map)

# Plot Proximities and Lines
for label, loc in proximities.items():
    Marker(loc, tooltip=label, icon=folium.Icon(color='green', icon='info-sign')).add_to(site_map)
    PolyLine(locations=[launch_site, loc], color='red', weight=2.5).add_to(site_map)

site_map
```
